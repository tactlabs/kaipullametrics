```python
from prettymetrics.clf import Classifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
data = pd.read_csv("../examples/dataset/kidney_stone_data.csv")
```

```python
data.head()
```

```python
data.stone_size.unique()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data['stone_size'] = le.fit_transform(data.stone_size)
data['treatment'] = le.fit_transform(data.treatment)
```

```python
data.head()
```

```python
X = data.drop(['success'], axis = 1)
y = data['success']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.3,random_state =123)
```

```python
clf = Classifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
```
