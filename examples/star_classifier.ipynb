```python
import prettymetrics
import pandas as pd 
```

```python
from prettymetrics.supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
df= pd.read_csv('../examples/datasets/Star3642_balanced.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

SpType_le = LabelEncoder()

df['SpType'] = SpType_le.fit_transform(df.SpType)
```

```python
df.head()
```

```python
X = df.drop(['TargetClass'], axis = 1)
Y = df['TargetClass']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15)
```

```python
X_train
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, Y_train, Y_test)

print(models)
```
